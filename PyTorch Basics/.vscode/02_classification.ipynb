# PyTorch NN Classification

- A classification problem involves predicting whether something is one thing or another
- Examples:
    1. **Binary Classification**: Target can be one of 2 options, like yes or no. Predict whether someone has a heart disease based on their health parameters
    2. **Multi-Class Classification**: Target can be one of 2+ options. Decide whether a photo is of food, a person, or a dog.
    3. **Multi-Label Classification**: Target can be assigned more than one options. Predict what categories should be assigned to a Wikipedia article.
- One of the most common ML algorithms along with regression
- Here we cover the following:
    1. Architecture of a classification NN
    2. Getting binary classification data ready
    3. Building a PyTorch Classification Model
    4. Fitting the model to the data (training)
    5. Making predictions and evaluating the model
    6. IMproving a Model 
    7. Non-linearity
    8. Replicating non-linear functions
    9. Putting it altogether with multiclass classification
