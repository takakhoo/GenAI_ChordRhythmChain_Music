```python
# Create a random tensor with shape (7,7)
import torch
tensor = torch.rand(7,7)
print(tensor)

#performing matrix multiplication on the tensor with another random tensor of shape (1,7)
tensor2 = torch.rand(1,7)
product = tensor @ tensor2.T
print(product)
```

```python
# Setting the random seed to 0 and do the same thing from above again
torch.manual_seed(0)
tensor1 = torch.rand(7,7)
tensor2 = torch.rand(1,7)
product = tensor1 @ tensor2.T
print(product)

torch.cuda.is_available()
```

```python
#Setting the GPU equivalent for random seed
torch.cuda.manual_seed(1234)
torch.cuda.is_available()
device = "cuda" if torch.cuda.is_available() else "cpu"

tensor_gpu1 = torch.rand(2,3)
tensor_gpu2 = torch.rand(2,3)
g1 = tensor_gpu1.to(device)
g2 = tensor_gpu2.to(device)

prod = g1 @ g2.T
print(prod, prod.shape)
```

```python
#Finding max and min of prod
print(prod.max(), prod.min(), prod.argmax(), prod.argmin())
```
